1. Número de defunciones fetales según el sexo.

```python
import _csv
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"]= (4,3)
plt.rcParams["figure.dpi"]= 150
lista = []
#Lectura del archivo .csv
with open('C:/Users/ALIENWARE/Documents/MEPN/Python/EDF_2019.csv',  encoding='utf-8-sig') as data:
    entrada = _csv.reader(data, delimiter = ';')
    lista = list(entrada)  
#Elimino cabecera.
lista.pop(0)
cont_h = 0
cont_m = 0
cont_i = 0
cont_s = 0
#Recorro la lista y realizo el conteo de los datos que cumplen la condición.
for linea in lista:
    if linea[7] =='Hombre':
        cont_h += 1
    elif linea[7] =='Mujer':
        cont_m += 1
    elif linea[7] =='Indeterminado':
        cont_i += 1
    elif linea[7] =='Sin información':
        cont_s += 1
#Imprimo resultado total.
print('Total individuos',len(lista))
#Grafico usando la librería panda y pyplot.
x = ['Hombres','Mujeres', 'Indeterminados', 'Sin Información' ]
y = [cont_h,cont_m, cont_i, cont_s]
plt.figure(figsize=(8, 5)) 
frecuencia_series = pd.Series(x)
frecuencia_series = pd.Series(y) 
ax = frecuencia_series.plot(kind='bar') 
ax.set_xticklabels(x, rotation=360) 
ax.set_title("Defunciones fetales en el año 2019 según el sexo") 
rects = ax.patches 
#Coloco anotación en cada barra.
for rect in rects: 
    y_value = rect.get_height()
    x_value = rect.get_x() + rect.get_width()/2 
    space = 0
    va = 'bottom' 
    if y_value < 0: 

     space *= -1 

     va = 'top' 

    label = "{:.1f}".format(y_value) 

    plt.annotate(
     label,      
     (x_value, y_value),   
     xytext=(0, space),   
     textcoords="offset points", 
     ha='center',   
     va=va) 
```

4. Porcentaje de defunciones fetales en la ciudad de Cuenca por parroquia.

```python
import _csv
import pandas as pd 
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"]= (4,3)
plt.rcParams["figure.dpi"]= 150
lista = []
#Lectura del archivo .csv
with open('C:/Users/ALIENWARE/Documents/MEPN/Python/EDF_2019.csv',  encoding='utf-8-sig') as data:
    entrada = _csv.reader(data, delimiter = ';')
    lista = list(entrada)  
#Elimino cabecera.
lista.pop(0)
cont_h = 0
cont_s = 0
cont_e = 0
cont = 0
#Defino condición, series y ejes
sexo = 'Mujer'
x = ['Huaynacápac','San Blas','El Batán']
#Recorro la lista y realizo el conteo de los datos que cumplen la condición.
for linea in lista:
    if linea[7] == sexo:
        if linea[2] == x[0]:
            cont_h += 1
            cont += 1
        elif linea[2] == x[1]:
            cont_s += 1
            cont += 1
        elif linea[2] == x[2]:
            cont_e += 1
            cont += 1
#Imprimo resultado total.
print('Total individuos', cont)
#Grafico.
y = [cont_h,cont_s, cont_e]
plt.title("Defunciones fetales en el año 2019 en Cuenca por parroquia, Mujeres")
plt.pie(y, labels=x, autopct="%0.1f %%")
plt.axis("equal")
plt.show()
```

5. Porcentaje de defunciones fetales cuya madre es menor a las 20 años.

```python
import _csv
import pandas as pd 
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"]= (4,3)
plt.rcParams["figure.dpi"]= 150
lista = []
#Lectura del archivo .csv
with open('C:/Users/ALIENWARE/Documents/MEPN/Python/EDF_2019.csv',  encoding='utf-8-sig') as data:
    entrada = _csv.reader(data, delimiter = ';')
    lista = list(entrada)  
#Elimino cabecera.
lista.pop(0)
cont_1 = 0
cont_2 = 0
cont_3 = 0
cont = 0
#Defino series y ejes
x = ['Edad en años: 10 - 12','Edad en años: 13 - 15','Edad en años: 16 - 19']
for linea in lista:
    if int(linea[28]) >=10 and int(linea[28]) <= 12:
        cont_1 += 1
        cont += 1
    elif int(linea[28]) >=13 and int(linea[28]) <= 15:
        cont_2 += 1
        cont += 1
    elif int(linea[28]) and int(linea[28]) < 20:
        cont_3 += 1
        cont += 1
#Imprimo resultado total.
print('Total individuos', cont)
#Grafico.
y = [cont_1,cont_2, cont_3]
plt.title("Defunciones fetales en el año 2019 cuya madre tenía menos de 20 años")
plt.pie(y, labels=x, autopct="%0.1f %%")
plt.axis("equal")
plt.show()
```
